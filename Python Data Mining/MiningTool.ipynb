# Imports needed for data manipulation & visualization

```python
import csv
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
```

# Reading Sales CSV file:

```python
file = pd.read_csv('All_sales.csv')
```

# Select Sample size from Sales

```python
df = pd.DataFrame(file)
```

```python
month1 = df.head(89)
```

```python
month12 = df.tail(112)
```

```python
m1SALES = sum(month1['SALES QUANTITY'])
```

```python
m12SALES = sum(month12['SALES QUANTITY'])
```

# Plotting Sales Quantity for first and last month 

```python
data = [m1SALES, m12SALES]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
period = ['First Month', 'Last Month']
ax.bar(period,data)
plt.xlabel("PERIOD")
plt.ylabel("SALES QUANTITY")
percChange = (m12SALES - m1SALES) / m1SALES *100

plt.show()
print(round(percChange, 1),'percent increase in items sold')
```
